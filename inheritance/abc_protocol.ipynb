```python
from typing import Protocol, runtime_checkable

@runtime_checkable
class IAssetPair(Protocol):
    def some_f(self)->str:
        ...
    def __str__(self):
        return "FROM PROTOCOL"

class X(IAssetPair):
    def some_f(self)->str:
        return "X"
    
class Y:
    def some_f(self)->str:
        return "Y"
```

```python
x = X()
print("isinstance(x, IAssetPair):", isinstance(x, IAssetPair))
print("str(x):", str(x))
print("x.some_f():", x.some_f())
```

```python
y = Y()
print("isinstance(y, IAssetPair):", isinstance(y, IAssetPair))
print("str(y):", str(y))
print("y.some_f():", y.some_f())
```
