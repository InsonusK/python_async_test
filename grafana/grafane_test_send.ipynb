```python
from dotenv import load_dotenv
import os 
env_path = os.path.join(".", "env","secrets","dev.secret")
load_dotenv(env_path)
_url = os.getenv("INFLUXDB_HOST")
_token = os.getenv("INFLUXDB_API_TOKEN")
_org = os.getenv("INFLUXDB_ORG")
_bucket = os.getenv("INFLUXDB_BUCKET")
```

```python
# Создайте клиент для подключения
from influxdb_client.client.influxdb_client import InfluxDBClient


__client = InfluxDBClient(url=_url, token=_token, org=_org)
# Получите объект для записи данных
__write_api = __client.write_api()
```

```python
from datetime import datetime,timedelta
import random
from influxdb_client.client.write.point import Point

start_date = datetime.now()
for i in range(1000):
    cur_date = start_date + timedelta(seconds=i*60)
    #int_tag = 5+random.randint(0,10)
    #float_tag = round(25.0+(random.randint(0,5000)/100),2)
    bool_tag = bool(random.randint(0,1))
    str_tag = "value"+str(random.randint(1,5))
    value1 = round(random.randint(0,10000)/100,2)
    value2 = round(random.randint(0,5000)/100,)
    point=Point("test_metric3")\
        .time(cur_date)\
        .tag("str_tag", str_tag).tag("bool_tag", bool_tag)\
        .field("field1", value1).field("field2", value2)
        #.tag("float_tag", float_tag ).tag("int_tag", int_tag)\
    __write_api.write(bucket=_bucket, org=_org, record=point)
```
